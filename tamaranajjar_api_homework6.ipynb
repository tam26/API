```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

from pprint import pprint
from api_keys import api_key
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []


lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

        
len(cities)
```

```python
# Save config information.

city_list = cities[0]
base_url = 'http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid='+ api_key+"&q="

url = base_url + city_list

weather_response = (requests.get(url)).json()

pprint(weather_response)
```

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

for city in cities:
    
    try:
        weather_response_all = requests.get(f"{base_url}{city}").json()
        city_record = weather_response_all['name']
        print(f"Processing Record {record} : {city_record}")  
        
        city_name.append(weather_response_all['name'])
        cloudiness.append(weather_response_all["clouds"]["all"])
        country.append(weather_response_all["sys"]["country"])
        date.append(weather_response_all["dt"])
        humidity.append(weather_response_all["main"]["humidity"])
        lat.append(weather_response_all["coord"]["lat"])
        lng.append(weather_response_all["coord"]["lon"])
        max_temp.append(weather_response_all["main"]["temp_max"])
        wind_speed.append(weather_response_all["wind"]["speed"])
        
        print(f"{base_url}{city}")
        
        record += 1
    
    except:
        print(weather_response_all)
       
    print("-"*20)
```

```python
weather_summary = {"City" : city_name,
                   "Cloudiness" : cloudiness,
                   "Country" : country,
                   "Date" : date,
                   "Humidity" : humidity,
                   "Latitude" : lat,
                   "Longitude" : lng,
                   "Max Temp" : max_temp,
                   "Wind Speed" : wind_speed
}

weather_df = pd.DataFrame(weather_summary)
weather_df.to_csv("cities.csv", encoding='utf-8', index=False)

print(weather_df.count())
weather_df.head()
```

```python
plt.scatter(weather_df["Latitude"], weather_df["Max Temp"], marker = "o", s = 35, edgecolor = "black", alpha = 0.8)

plt.grid()
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title("City Latitude vs. Max Temperature (08/22/2018)")

plt.savefig("Lat v. Max Temp.png")

plt.show()
```
